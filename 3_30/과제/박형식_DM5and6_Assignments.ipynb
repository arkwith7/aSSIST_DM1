## 과제 2. 휴대기기 유료앱 사용 및 인앱결재에 미치는 요인 분석

- Logistic Regression 이용
- NB 이용
- Decision Trees 이용
- Random Forest 이용
- XG Boost 이용
- 모형간 비교

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

```

```python

# 데이터 전처리
# OTT 사용여부(1있음, 2없음) 컬럼을 기준으로 OTT 사용자와 비사용자를 분리
ott_user = media[media['OTT'] == 1]
ott_non_user = media[media['OTT'] == 2]
```

```python
# OTT 사용여부 컬럼을 기준으로 사용자는 1, 비사용자는 0으로 변환
ott_user['OTT'] = 1
ott_non_user['OTT'] = 0
```

```python
# 사용자와 비사용자의 비율을 맞추기 위해 사용자와 비사용자의 데이터를 1:1로 맞춤
# ott_user = ott_user.sample(n=len(ott_non_user), random_state=42)
ott_user = ott_user.sample(n=len(ott_non_user), replace=True, random_state=42)
ott_non_user = ott_non_user.sample(n=len(ott_user), random_state=42)
```

```python
# 사용자와 비사용자의 데이터를 합쳐 전체 데이터 생성
data = pd.concat([ott_user, ott_non_user])
```

```python
# 전체 데이터에서 OTT 사용여부 컬럼을 제외한 나머지 컬럼을 독립변수로 설정
X = data.drop('OTT', axis=1)

# X 데이터에서 오브젝트 타입을 제외한 컬럼만 선택
X = X.select_dtypes(exclude='object')

# # 공백을 NaN으로 변환
# X.replace(' ', np.nan, inplace=True)

# # 수치형 데이터로 변환 (오류를 방지하기 위해 적절한 변환 적용)
# X = X.apply(pd.to_numeric, errors='coerce')

# # 결측치를 평균값으로 대체
# imputer = SimpleImputer(strategy='mean')
# X = imputer.fit_transform(X)
```

```python
# X 데이터를 확인
print(X.shape)
print(X[:5])
```

```python
# OTT 사용여부 컬럼을 종속변수로 설정
y = data['OTT']
```

```python
# 독립변수와 종속변수를 나누어 학습용 데이터와 테스트용 데이터로 나눔
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# 학습용 데이터와 테스트용 데이터를 표준화
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# 모델링
# Logistic Regression
lr = LogisticRegression()
lr.fit(X_train, y_train)
lr_pred = lr.predict(X_test)
lr_accuracy = accuracy_score(y_test, lr_pred)
lr_cm = confusion_matrix(y_test, lr_pred)
lr_cr = classification_report(y_test, lr_pred)

```

```python

# Naive Bayes
nb = GaussianNB()
nb.fit(X_train, y_train)
nb_pred = nb.predict(X_test)
nb_accuracy = accuracy_score(y_test, nb_pred)
nb_cm = confusion_matrix(y_test, nb_pred)
nb_cr = classification_report(y_test, nb_pred)

```

```python
# Decision Trees
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
dt_pred = dt.predict(X_test)
dt_accuracy = accuracy_score(y_test, dt_pred)
dt_cm = confusion_matrix(y_test, dt_pred)
dt_cr = classification_report(y_test, dt_pred)
```

```python

# XG Boost
xgb = XGBClassifier()
xgb.fit(X_train, y_train)
xgb_pred = xgb.predict(X_test)
xgb_accuracy = accuracy_score(y_test, xgb_pred)
xgb_cm = confusion_matrix(y_test, xgb_pred)
xgb_cr = classification_report(y_test, xgb_pred)
```

```python
# 모델간 비교
models = ['Logistic Regression', 'Naive Bayes', 'Decision Trees', 'XG Boost']
accuracy = [lr_accuracy, nb_accuracy, dt_accuracy, xgb_accuracy]
cm = [lr_cm, nb_cm, dt_cm, xgb_cm]
cr = [lr_cr, nb_cr, dt_cr, xgb_cr]

model_comparison = pd.DataFrame({'Model': models, 'Accuracy': accuracy, 'Confusion Matrix': cm, 'Classification Report': cr})
model_comparison
```

#### 결론
- Logistic Regression, Naive Bayes, Decision Trees, XG Boost를 이용하여 OTT 사용여부를 분석한 결과, Decision Trees의 정확도가 0.79로 가장 높았습니다.

```python

# Feature Importance
# - Logistic Regression의 결과를 시각화하여 확인합니다.
import matplotlib.pyplot as plt
import numpy as np

# 로지스틱 회귀 모델의 계수를 가져옵니다.
coefficients = lr.coef_[0]

# 계수의 절대값을 기준으로 특성 중요도를 시각화합니다.
feature_importance = pd.Series(coefficients, index=X.columns).sort_values(key=abs)

plt.figure(figsize=(10, 8))
feature_importance.plot(kind='barh')
plt.title('Feature Importance in Logistic Regression')
plt.xlabel('Coefficient Value')
plt.ylabel('Features')
plt.show()
```

```python
# - Naive Bayes의 결과를 시각화하여 확인합니다.
# 나이브 베이즈 모델의 특성 중요도를 시각화합니다.
feature_importance = pd.Series(nb.theta_[0], index=X.columns).sort_values()

plt.figure(figsize=(10, 8))
feature_importance.plot(kind='barh')
plt.title('Feature Importance in Naive Bayes')
plt.xlabel('Feature Value')
plt.ylabel('Features')
plt.show()
```
