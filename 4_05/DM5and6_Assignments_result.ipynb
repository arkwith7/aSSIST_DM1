```python
data_file = 'data/mediapanel_2019.csv'
media = pd.read_csv(data_file, index_col='pid')
media.head(5)
```

##  과제 1. OTT 사용율에 미치는 영향 분석

```python
# 데이터 확인
media.info()
```

```python
for col in media.columns:
    value_counts = media[col].value_counts()
    print(f"{col}:\n{value_counts}\n")
```

```python
# Target 
y = media['OTT']
y[y==2]=0
```

```python
y.value_counts()
```

```python
# dummy variables 세트 만들기
# Binary variables
binary = ['gender','job','relig','pet','cell','featurephone','smartphone','kidsphone',
         'tablet','wearable','payapp','inapp_pay','Newspaper','email','messanger','blog',
         'cloud','SNS']
X1 = media[binary]

# convert non-binary categorical variables to dummies
nonbinary = ['school','marriage','smart_or_not']
X2 = media[nonbinary]

school_dum = pd.get_dummies(X2['school'], prefix='school')
school_dum.drop(school_dum.columns[0], axis=1, inplace=True)

marriage_dum = pd.get_dummies(X2['marriage'], prefix='marriage')
marriage_dum.drop(marriage_dum.columns[0], axis=1, inplace=True)

billtype_dum = pd.get_dummies(X2['smart_or_not'], prefix='billtype')
billtype_dum.drop(billtype_dum.columns[0], axis=1, inplace=True)

X1 = pd.concat([X1, school_dum, marriage_dum, billtype_dum], axis=1)

# replace 2 to 0, and ' ' to 0
X1 = X1.replace(2,0)
X1 = X1.replace(' ',0)
X1.head(3)
```

```python
X1.info()
```

```python
X1 = X1.astype('int64')
```

```python
X1.info()
```

```python
# continuous variable set 만들기
conts = ['age','income','theater','theaterpay','play','playpay',]
X3 = media[conts]
X3 = X3.replace(' ',0).astype('int64')
X3.info()
X3.head(3)
```

```python
# 최종 데이터 세트
X = pd.concat([X1, X3], axis=1)
```

```python
# 트레이닝/테스트 분할
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, 
                                                    test_size=0.3, random_state=123)
```

```python
# NB를 위해 연속/더비 분할
X_train_cat = X_train[X1.columns]
X_train_cont = X_train[X3.columns]
X_test_cat = X_test[X1.columns]
X_test_cont = X_test[X3.columns]
```

```python
# 필요한 함수 불러오기
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import CategoricalNB, GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import roc_curve, auc, roc_auc_score
from sklearn import metrics
```

```python
# 로지스틱 리그레션
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_pred_logreg = logreg.predict(X_test)
```

```python
# NB
cnb = CategoricalNB()
gnb = GaussianNB()
cnb.fit(X_train_cat, y_train)
gnb.fit(X_train_cont, y_train)

# 연속변수 추정
y_cont_prob = gnb.predict_proba(X_test_cont)

# 범주형변수 추정
y_cat_prob = cnb.predict_proba(X_test_cat)

# 추정차 합치기
y_pred_prob = y_cont_prob * y_cat_prob
y_pred_nb = np.argmax(y_pred_prob, axis=1)
```

```python
# DT
dt = DecisionTreeClassifier(max_depth=5)
dt.fit(X_train, y_train)
y_pred_dt = dt.predict(X_test)
```

```python
# RF
rf = RandomForestClassifier(n_estimators=1000, max_depth = 5)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
```

```python
#XGBoost
xgb = XGBClassifier(
    objective="binary:logistic",  # 분류 문제에 사용하는 목표 함수
    eval_metric="logloss",        # 평가 메트릭스
    max_depth=3,                  # 트리의 최대 깊이
    learning_rate=0.1,            # 학습률
    n_estimators=100,             # 트리의 개수
    random_state=42               # 랜덤 시드
)
xgb.fit(X_train, y_train)
y_pred_xgb = xgb.predict(X_test)
```

```python
# Accuracy
accuracy_lr = metrics.accuracy_score(y_test, y_pred_logreg)
accuracy_nb = metrics.accuracy_score(y_test, y_pred_nb)
accuracy_dt = metrics.accuracy_score(y_test, y_pred_dt)
accuracy_rf = metrics.accuracy_score(y_test, y_pred_rf)
accuracy_xgb = metrics.accuracy_score(y_test, y_pred_xgb)
print("accuracies of each models are:", accuracy_lr, accuracy_nb,accuracy_dt,accuracy_rf, accuracy_xgb)
```

```python
# other evaluations
print(metrics.classification_report(y_test, y_pred_logreg))
print(metrics.classification_report(y_test, y_pred_nb))
print(metrics.classification_report(y_test, y_pred_dt))
print(metrics.classification_report(y_test, y_pred_rf))
print(metrics.classification_report(y_test, y_pred_xgb))
```

```python
# AUC
auc_lr=roc_auc_score(y_test, y_pred_logreg)
auc_nb=roc_auc_score(y_test, y_pred_nb)
auc_dt=roc_auc_score(y_test, y_pred_dt)
auc_rf=roc_auc_score(y_test, y_pred_rf)
auc_xgb=roc_auc_score(y_test, y_pred_xgb)
print(auc_lr, auc_nb, auc_dt, auc_rf, auc_xgb)
```
