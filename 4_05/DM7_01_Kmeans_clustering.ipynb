# K평균 군집분석

sci-kit Learn에서 제공하는 iris 데이터세트를 가지고 군집분석을 해 봅니다.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## iris데이터 설명

붓꽃(iris)의 품종(iris.target), 꽃받침(sepal)의 길이 및 넓이, 그리고 꽃임(petal)의 길이와 넓이를 조사한 데이터


```python
# iris데이터 불러오기
from sklearn.datasets import load_iris
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target_names[iris.target]
```

```python
df.head()
```

```python
df['target'].value_counts()
```

## Clustering은 Unsupervised learning 입니다. 필요없는 target변수는 제거하고 분류를 해 봅니다.

* 우선 필요한 데이터만 뽑습니다.
* 임의의 K값을 통해 K평균 군집분석을 해 봅니다.

```python
from sklearn.cluster import KMeans # K-means 군집분석 함수를 불러옵니다.
X = iris.data

# 2개로 해 봅니다.
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)

print(kmeans)
```

```python
print(X)
```

## 정확도를 계산해 봅니다.
실제 타겟변수가 없는 데이터를 통한 K평균 분석에서는 필요없는 작업입니다.

```python
y_test = iris.target
y_pred = kmeans.labels_

print("y_test is:", y_test)
print("y_pred is:", y_pred)
```

## 3개로 해 봅니다.

```python
# 3개로 해 봅니다.
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)

y_pred = kmeans.labels_

print("y_test is:", y_test)
print("y_pred is:", y_pred)
```

```python
# visualize 해 봅니다.
from sklearn.decomposition import PCA

# Perform PCA for visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

print(pca.components_)
```

```python
# Visualize the clustering result
colors = np.array(['#ff7f0e', '#1f77b4', '#2ca02c'])
plt.scatter(X_pca[:, 0], X_pca[:, 1], s=50, color=colors[y_pred])
plt.scatter(kmeans.cluster_centers_[:, 1], kmeans.cluster_centers_[:, 2], s=200, marker='*', color='red')
plt.xlabel('PC 1')
plt.ylabel('PC 2')
plt.title('K-means Clustering Result')
plt.show()
```

```python
# 4개로 해 봅니다.
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)

y_pred = kmeans.labels_

print("y_test is:", y_test)
print("y_pred is:", y_pred)
```

## "elbow method"를 활용하여 최적의 수 K를 구해봅니다.

```python
# Define range of K values to try
k_values = range(1, 10)

ssd = []
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(X)
    ssd.append(kmeans.inertia_)  # kmeans.inertia_ 는 팔꿈치 법에 사용하는 Sum of Squared Distance를 의미합니다.

# Plot the results to find the "elbow" point
plt.plot(k_values, ssd, 'bx-')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Sum of Squared Distances')
plt.title('Elbow Method for Optimal K (Iris dataset)')
plt.show()
```

## Kmeans++분석을 해 봅니다.

```python
kmeans = KMeans(n_clusters=3, init='k-means++') # 기존 K평균과 동일하지만, 옵션만 init='k-means++' 넣어주면 됩니다.
y_pred = kmeans.fit_predict(X)

```

```python
# Visualize the clustering result
colors = np.array(['#ff7f0e', '#1f77b4', '#2ca02c'])
plt.scatter(X_pca[:, 0], X_pca[:, 1], s=50, color=colors[y_pred])
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, marker='*', color='red')
plt.xlabel('PC 1')
plt.ylabel('PC 2')
plt.title('K-means++ Clustering Result')
plt.show()
```
