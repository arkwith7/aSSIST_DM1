```python
import networkx as nx

# Creating the graph
G = nx.DiGraph()
```

```python
# Adding nodes with their respective categories as node attributes for later grouping if needed
nodes = {
    "실손의료보험(4세대)": {"category": "정의"},
    "기본형": {"category": "보장 유형"},
    "특별약관": {"category": "보장 유형"},
    "3대 비급여": {"category": "보장 유형"},
    "상해급여": {"category": "기본형"},
    "질병급여": {"category": "기본형"},
    "상해비급여": {"category": "특별약관"},
    "질병비급여": {"category": "특별약관"},
    "상급병실료 차액보장": {"category": "특별약관"},
    "도수·체외충격파·증식 치료": {"category": "3대 비급여"},
    "주사료": {"category": "3대 비급여"},
    "자기공명영상진단": {"category": "3대 비급여"}
}

# Adding nodes to the graph
for node, attrs in nodes.items():
    G.add_node(node, **attrs)
```

```python
# Adding edges
edges = [
    ("실손의료보험(4세대)", "기본형"),
    ("실손의료보험(4세대)", "특별약관"),
    ("실손의료보험(4세대)", "3대 비급여"),
    ("기본형", "상해급여"),
    ("기본형", "질병급여"),
    ("특별약관", "상해비급여"),
    ("특별약관", "질병비급여"),
    ("특별약관", "상급병실료 차액보장"),
    ("3대 비급여", "도수·체외충격파·증식 치료"),
    ("3대 비급여", "주사료"),
    ("3대 비급여", "자기공명영상진단")
]

# Adding edges to the graph
G.add_edges_from(edges)
```

```python
# Plotting the graph
plt.figure(figsize=(14, 8))
pos = nx.spring_layout(G)  # positions for all nodes
nx.draw(G, pos, with_labels=True, node_size=7000, node_color="skyblue", font_size=9, font_weight='bold', font_family='sans-serif')
plt.title("실손의료보험(4세대) 지식 그래프")
plt.show()
```
