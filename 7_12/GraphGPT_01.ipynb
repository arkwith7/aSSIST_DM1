```python
import networkx as nx

# Creating the graph
G = nx.DiGraph()
```

```python
# Adding nodes with their respective categories as node attributes for later grouping if needed
nodes = {
    "(4)": {"category": ""},
    "": {"category": " "},
    "": {"category": " "},
    "3 ": {"category": " "},
    "": {"category": ""},
    "": {"category": ""},
    "": {"category": ""},
    "": {"category": ""},
    " ": {"category": ""},
    " ": {"category": "3 "},
    "": {"category": "3 "},
    "": {"category": "3 "}
}

# Adding nodes to the graph
for node, attrs in nodes.items():
    G.add_node(node, **attrs)
```

```python
# Adding edges
edges = [
    ("(4)", ""),
    ("(4)", ""),
    ("(4)", "3 "),
    ("", ""),
    ("", ""),
    ("", ""),
    ("", ""),
    ("", " "),
    ("3 ", " "),
    ("3 ", ""),
    ("3 ", "")
]

# Adding edges to the graph
G.add_edges_from(edges)
```

```python
import matplotlib.pyplot as plt
import networkx as nx
import matplotlib.font_manager as fm

#     (: '/usr/share/fonts/truetype/nanum/NanumGothicCoding.ttf')
font_path = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'  #    
font_prop = fm.FontProperties(fname=font_path)

# G = nx.DiGraph()  #   
#    

plt.figure(figsize=(14, 8))
pos = nx.spring_layout(G, k=0.3, iterations=20)  # k      
nx.draw(G, pos, with_labels=True, font_size=14, node_color='skyblue', node_size=700, font_family=font_prop.get_name(), font_weight='bold')
plt.title("(4)  ", fontproperties=font_prop)
plt.show()
```
