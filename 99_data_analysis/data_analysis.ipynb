```python
# 필요한 라이브러리를 불러옵니다.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# matplotlib 그래프를 위한 설정
plt.rcParams['figure.figsize'] = [10, 6]
plt.rcParams['font.size'] = 12
plt.rcParams['font.family'] = 'NanumGothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 데이터를 로드합니다.
file_path = './youtube_channel_comments_data_20240530_084814.csv'
data = pd.read_csv(file_path)

# 데이터의 처음 몇 줄을 출력하여 확인합니다.
print(data.head())
```

```python
# 데이터의 기본 정보를 출력합니다.
print(data.info())
```

```python
# 결측치의 수를 확인합니다.
print(data.isnull().sum())
```

```python
# 기술 통계를 출력하여 데이터를 더 잘 이해합니다.
print(data.describe())
```

```python
# 각 비디오별 댓글 수를 시각화합니다.
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='videoId')
plt.title('댓글 수에 따른 비디오 ID')
plt.xticks(rotation=45)
plt.show()
```

```python
# 댓글의 'likeCount' 분포를 시각화합니다.
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='likeCount', bins=30, kde=True)
plt.title('댓글의 좋아요 수 분포')
plt.show()
```
