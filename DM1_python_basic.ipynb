기본적인 계산입니다. print() 함수는 괄호안의 연산결과를 화면에 보여주는 함수입니다.

```python
# Sample code for arithmetic operationsß

# Addition
print(3 + 5) # Output: 8

# Subtraction
print(5 - 3) # Output: 2

# Multiplication
print(3 * 5) # Output: 15

# Division
print(10 / 2) # Output: 5.0

# Floor division
print(10 // 3) # Output: 3

# Modulus
print(10 % 3) # Output: 1

# Boolean Logic Operator
print(3==4) # Output: False
print (2 <= 5) #Output: True
```

숫자뿐만 아니라 문자형변수(String)를 사용해서도 다양한 연산을 할 수 있습니다.
대표적으로 두 문자변수를 연결하거나, 여러번 반복하거나, 문자열 중 특별한 문자만 골라내는 연산을 수행할 수 있습니다.

```python
# Sample code for string operations

# Concatenation
first_name = "John"
last_name = "Doe"
full_name = first_name + " " + last_name
print(full_name) # Output: John Doe

# Repetition
print("Hello " * 3) # Output: Hello Hello Hello 

# Slicing
word = "Python"
print(word[0:2]) # Output: Py
```

파이썬 데이터 타입은 정수형(integer), 유리수(float), 문자(string) 그리고 논리형(boolean)이 있습니다. 논리형은  19세기 수학자 Bool이 제안한 Boolean 대수논리를 기반으로 하는 논리연산 결과를 표시합니다. Boolean 대수논리는 일반적인 논리연산을 의미하며 참/거짓을 각각 0, 1로 표현하기도 합니다.


```python
# Sample code for type casting

# int()
print(int(3.14)) # Output: 3

# float()
print(float(3)) # Output: 3.0

# str()
print(str(3.14)) # Output: "3.14"

# bool()
print(bool(0)) # Output: False
```

조건문과 루프는 코딩의 처음이자 마지막이라 할 수 있습니다. 조건문과 루프에 익숙해져야지만 능수능란하게 코딩을 할 수 있습니다.
조건문은 해당 조건이 참/거짓임에 따라 다른 연산을 수행하도록 구성하는 방법이며, 루프는 특정 조건을 만족할 때 까지 반복하여 연산을 수행합니다.

```python
# If-Else Statements
x = 10
y = 20
if x > y:
  print("x is greater than y")
else:
  print("y is greater than x")
  
# For Loops
numbers = [1, 2, 3, 4, 5]
for number in numbers:
  print(number)
  
# While Loops
counter = 0
while counter < 5:
  print(counter)
  counter += 1
```

**함수와 함수정의**

함수는 기본적인 계산식이나 명령어를 미리 정의하여 불러오는 방법입니다. python언어에는 기본적으로 제공하는 다양한 operator 가 있습니다. 그러나, 조금 더 복잡한 계산을 하기 위해서는 미리 함수를 정의하는 것이 좋습니다.\
함수를 정의할 때는 def라는 함수를 사용하며 전체적인 문법은

def name_of_function(inputs):\
> operation\
> returns

이렇게 구성합니다.


```python
# Defining Functions
def greet(name):
  print("Hello, " + name)
  
# Calling Functions
greet("John Doe")
greet("Jane Doe")

# Returning Values
def add(x, y):
  return x + y
  
result = add(5, 10)
print(result)
```

```python
# Lists
# Creating lists
list1 = [1, 2, 3, 4, 5]
list2 = ['apple', 'banana', 'cherry']
list3 = [1, 2.3, "apple"]

# Accessing elements
print(list1[0]) # 1
print(list2[1]) # 'banana'
print(list3)
```

```python

# Dictionaries
# Creating dictionaries
dict1 = {'key1': 1, 'key2': 2, 'key3': 3}
dict2 = {'fruit1': 'apple', 'fruit2': 'banana', 'fruit3': 'cherry'}

# Accessing elements
print(dict1['key1']) # 1
print(dict2['fruit2']) # 'banana'

# Updating dictionaries
dict1['key1'] = 10
dict2['fruit4'] = 'date'

```

```python
# Sets
# Creating sets
set1 = {1, 2, 3, 4, 5}
set2 = {'apple', 'banana', 'cherry'}

# Accessing elements
print(1 in set1) # True
print('apple' in set2) # True

# Updating sets
set1.add(6)
set2.remove('banana')
print(set1)
print(set2)
```
